```python
from numpy import genfromtxt
trainset1 = genfromtxt('trainset.csv', delimiter=',')
testset1 = genfromtxt('testset.csv', delimiter=',')
testlabel = genfromtxt('testlabel.csv', delimiter=',')
trainlabel = genfromtxt('trainlabel.csv', delimiter=',')
```

```python
print(set(list(testlabel)))
```

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
"""# define example
data = ['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm', 'cold', 'warm', 'hot']
values = array(data)
print(values)
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded)"""
integer_encoded=trainlabel
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded_train_label = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded_train_label)
integer_encoded=trainlabel
# binary encode
integer_encoded=testlabel

integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded_test_label = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded_test_label)
```

```python
print(trainset1.shape,onehot_encoded_train_label.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(trainset1,onehot_encoded_train_label)
print(model.score(trainset1,onehot_encoded_train_label))
predictionsgnb=model.predict(testset1)
predictionsgnb = model.predict(testset1)
count=0
for i in range(len(predictionsgnb)):
    if predictionsgnb[i]==onehot_encoded_test_label[i]:
        count=count+1
print("accuracy=",(count*100)/len(onehot_encoded_test_label))
```

```python
a=[]
b=[]
from sklearn.model_selection import KFold
from sklearn.neural_network import MLPClassifier
kf = KFold(n_splits=10)
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
avg=0
for train_indices, test_indices in kf.split(trainset1):
    clf.fit(trainset1[train_indices], trainlabel[train_indices])
    #print(clf.score(trainset[test_indices], trainlabel[test_indices]))
    prediction=clf.predict(testset1)
    c=0
    for i in range(len(prediction)):
        if prediction[i]==testlabel[i]:
            c=c+1
            a.append(testlabel[i])
            
        else:
            b.append(testlabel[i])
            print("actual=",testlabel[i],"   predicted=",prediction[i])
            
    print(c) 
    res=c*100/len(testlabel)
    avg=avg+res
    print("accuracy=",res)
    break
print("avg accuracy=",avg/10)
```
